```python
# Tính tổng bình phương 
def sum_of_squares(s): 
   return sum(s_i * s_i 
          for s_i, s_i in zip(s, s))
# Định nghĩa hàm tính mean
def mean(s):
    return sum(s)/len(s)

# tính độ lệch
def deviation(s): 
   s_Mean = mean(s)
   return [s_i - s_Mean for s_i in s]

# tính phương sai 
def variance(s): 
   n = len(s) 
   d = deviation(s) 
   return sum_of_squares(d) / n
```

Accuracy Score

```python
#accuracy on test data
x_ts_predict = model.predict(x_ts)
test_data_accuracy = accuracy_score(y_ts, x_ts_predict)
print(test_data_accuracy)
```
